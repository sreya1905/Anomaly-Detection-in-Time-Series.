# 1. AMAZON

```python
#import libraries which are necessery

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read dataset 
df=pd.read_csv(r"Amazon.csv")
```

```python
#calls dataset
df
```

```python
#returns description of data in dataframe
df.describe()
```

```python
#prints information about the data
df.info()
```

```python
#checks if there is any null value
df.isnull()
```

```python
#checks if there is any null value and returns sum of it.
df.isnull().sum()
```

```python
#gets column names
df.columns
```

```python
#defines a list
c=['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
```

```python
#loop is given here to get boxplots of the given column datas to find outliers
for i in c:
    sns.boxplot(data=df[i])
    plt.show()
```

# Manage outliers

```python
sns.boxplot(df["Volume"])
```

```python
Q1=np.percentile(df["Volume"],25,interpolation="midpoint")
Q2=np.percentile(df["Volume"],50,interpolation="midpoint")
Q3=np.percentile(df["Volume"],75,interpolation="midpoint")
IQR=Q3-Q1
up_lim=Q3+1.5*IQR
outlier=[]
for x in df["Volume"]:
    if(x>up_lim):
        outlier.append(x) 
index1 = (df['Volume']>up_lim)
index1 = df.loc[index1].index
df.loc[list(index1),'Volume']=up_lim
sns.boxplot(df["Volume"])
```

```python
for i in c:
    sns.boxplot(data=df[i])
    plt.show()
```

```python
df.isnull().sum()
```

# Visualization

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.corr(),annot=True,cmap="crest")
```

```python
amazon_c=df
amazon_c.to_csv("amazon c.csv",index=False)
```
